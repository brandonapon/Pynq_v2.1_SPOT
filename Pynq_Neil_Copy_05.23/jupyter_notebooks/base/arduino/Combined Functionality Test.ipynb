```python
from pynq.overlays.base import BaseOverlay
import time
base = BaseOverlay("base.bit")
```

```python
from pynq.lib.arduino import Arduino_SPOT

tmp = Arduino_SPOT(base.ARDUINO,Arduino_SPOT)
```

```python
tmp.play_sequence([16,-100,16,-100,16,-50,16])
time.sleep(20)
tmp.stop()
```

```python
i = 0
while i <= 60:
    print(i, " seconds")
    print(tmp.get_gyro())
    print(tmp.get_compass())
    print(tmp.get_accl())
    print(tmp.get_euler())
    print(tmp.get_temperature())
    time.sleep(1)
    i = i + 1
```

```python
i = 0
while i <= 60:
    print(i, " seconds")
    [ex, ey, ez] = tmp.get_euler()
    print(ex)
    time.sleep(1)
    i = i + 1
    if(ex > 350.0) :
        tmp.play_sequence([16])
        tmp.poll()
        time.sleep(1)
        i = i + 1
    if(ex < 10.0) :
        tmp.play_sequence([16])
        tmp.poll()
        time.sleep(1)
        i = i + 1
        
tmp.stop()
    
```
