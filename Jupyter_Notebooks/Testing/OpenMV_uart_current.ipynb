```python
from pynq.overlays.base import BaseOverlay
base = BaseOverlay("base.bit")

from pynq.lib.arduino import Arduino_CAM_UART
pynq = Arduino_CAM_UART(base.ARDUINO)
import time

WRITE = 0x02
READ = 0x03
TRANSACTION = 0x04
SIMPLE = 0x05
STREAM = 0x06
STREAM_DDR = 0x07
STREAM_PIXEL = 0x08
FLOW = 0x09
CONSECUTIVE = 0x0a
LOW_LEVEL = 0x0b
```

```python
#testing stream small
stream_size = 256

pynq.microblaze.write_mailbox(0, stream_size)
pynq.microblaze.write_blocking_command(STREAM)
read_in = pynq.microblaze.read_mailbox(0,stream_size+2)

print('read_count =', read_in[stream_size])
print('stream_size =', read_in[stream_size+1])

for index in range(0,stream_size):
#     print(chr(val) + " " + str(index))
    print(int(read_in[index]))
```

```python
#testing stream w/ ddr
stream_size = 1024

buf0 = pynq.buf_manager.cma_alloc(stream_size, data_type="uint8_t")
buf1 = pynq.buf_manager.cma_get_buffer(buf0, stream_size)
phy_addr = pynq.buf_manager.cma_get_phy_addr(buf0)

pynq.microblaze.write_mailbox(0, stream_size)
pynq.microblaze.write_mailbox(4, phy_addr)
pynq.microblaze.write_blocking_command(STREAM_DDR)
read_in = pynq.microblaze.read_mailbox(0,3)

print('read_count =', read_in[0])
print('stream_size =', read_in[1])

for index in range (0, stream_size):
    print(int((buf1[index].hex()), 16))
```
